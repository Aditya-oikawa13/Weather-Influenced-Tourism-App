```python
from urllib.parse import urlencode
import requests
```

```python
def extract_lat_long(address, data_type = 'json'):
    endpoint = f'https://maps.googleapis.com/maps/api/geocode/{data_type}'
    params = {'address': address, 'key': api_key}
    url_params = urlencode(params)
    url = f'{endpoint}?{url_params}'
    r = requests.get(url)
    if r.status_code not in range(200,299):
        return{}
    latlong = {}
    try:
        latlong =  r.json()['results'][0]['geometry']['location']
    except:
        pass
    return latlong.get('lat'), latlong.get('lng')
```

```python
lat, long = extract_lat_long('Tokyo, japan')
```

```python
def getting_places(latitude, longitude, radius = 2000, next_page_token = None):
    base_endpoint_places = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    params = {
    'key' : api_key,
    'location': f'{lat},{long}',
    'keyword' : ['attractions'],
    'radius': radius,
    'next_page_token': next_page_token
    }
    encoded_param = urlencode(params)
    places_url = f'{base_endpoint_places}?{encoded_param}'
    r = requests.get(places_url)
    places = r.json()
    return places
```

```python
places_data = getting_places(latitude=lat, longitude=long)
```

```python
places_data
```

```python
def places_info(places_data):
    info_list = []
    for result in places_data['results']:
        info_dict = {}
        info_dict['name'] = result['name']
        if 'rating' in result:
            info_dict['rating'] = result['rating']
        else:
            info_dict['rating'] = 'N/A'
        if 'vicinity' in result:
            info_dict['vicinity'] = result['vicinity']
        else:
            info_dict['vicinity'] = 'N/A'
        info_list.append(info_dict)
        info_dict['keywords'] = result['types']
    return info_list


        
        

```

```python
final_places_list = places_info(places_data)
```
